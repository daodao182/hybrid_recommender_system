```python
#importing libraries
import pandas as pd
```

```python
pd.set_option("display.max_columns", 20)
pd.pandas.set_option('display.width', 300)
```

```python
#Data Processing
#reading dataset "movie"
#27K dataset
movie = pd.read_csv("movie.csv")
#20M dataset
rating = pd.read_csv("rating.csv")
```

```python
movie.head()
```

```python
rating
```

```python
#merging two dataset into one dataframe
df = movie.merge(rating, how="left", on="movieId")
df.shape
```

```python
#https://www.skytowner.com/explore/pandas_dataframe_nunique_method
#How many unique movie title
df["title"].nunique()
```

```python
#counts number of user watch for the movie.
#Descending order
df["title"].value_counts()
```

```python
#constructing dataframe for the above movie watch
comment_counts = pd.DataFrame(df["title"].value_counts())
comment_counts
```

```python
#categorizing movies with comment count less than 1500 as the rare movies
rare_movies = comment_counts[comment_counts["title"] <= 1500].index
rare_movies
```

```python
#refining movies
#removing the rare movies from the dataframe
mostly_watched_movies = df[~df["title"].isin(rare_movies)] 
mostly_watched_movies
```

```python
#Number of unique movie titles 
mostly_watched_movies["title"].nunique()
```

```python
#pivot table for mostly watched movies
user_movie_df = mostly_watched_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
user_movie_df
```

```python
#checking movies watched by a random user
#For user-based recommendation
random_user = int(pd.Series(user_movie_df.index).sample(1, random_state=50).values)
random_user_df = user_movie_df[user_movie_df.index == random_user]
#One random users movie selection among 2449
random_user_df
```

```python
#converting the random user into a list
movies_watched = random_user_df.columns[random_user_df.notna().any()].to_list()
len(movies_watched)
```

```python
#extracting users that watched the same movie into a new dataframe
pd.set_option("display.max_columns", 8)
movies_watched_df = user_movie_df[movies_watched]
movies_watched_df
```

```python
#number of movies watched by users that matches the random user
user_movie_count = movies_watched_df.T.notnull().sum()
user_movie_count
```

```python
user_movie_count = user_movie_count.reset_index()
user_movie_count.columns = ["userId", "movie_count"]
user_movie_count
```

```python
# check = user_movie_count[user_movie_count["movie_count"] > 15].sort_values("movie_count", ascending=False)
# check
# user_movie_count[user_movie_count["movie_count"] == len(movies_watched)].count()
```

```python
#Identify the users who are most similar to the user to be suggested.
#60% of movies_watched = 16.2
percent_rand_user = len(movies_watched) * 60 / 100
percent_rand_user
```

```python
#number of users who watched 60% of the same movie as the random user
users_same_movies = user_movie_count[user_movie_count["movie_count"] > percent_rand_user]["userId"]
users_same_movies.count()
```

```python
#dataframe consisting users who watched the same movies as the random_user
final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movies.index)],random_user_df[movies_watched]])
final_df
```

```python
# final_df.T.corr()
```

```python
#finding correlations between users
corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
corr_df = pd.DataFrame(corr_df, columns=["correlation"])
corr_df.index.names = ["user_id_1", "user_id_2"]
corr_df = corr_df.reset_index()
corr_df.head(20)
```

```python
# 65% corerelated with random user
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["correlation"] >= 0.65)][["user_id_2", "correlation"]].reset_index(drop=True)
```

```python
#the most-correkated on top
top_users = top_users.sort_values(by="correlation", ascending=False)
```

```python
top_users.rename(columns={"user_id_2":"userId"}, inplace=True)
top_users
```

```python
#merging the rating dataset and the correlation
top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how="inner")
top_users_ratings
```

```python
top_users_ratings = top_users_ratings[top_users_ratings["userId"] != random_user]
top_users_ratings
```

```python
# top_users_ratings["userId"].unique()
# top_users_ratings
```

```python
#Weighted Average Recommendation Score : rating and correlation
top_users_ratings["weighted_rating"] = top_users_ratings["correlation"] * top_users_ratings["rating"]
top_users_ratings
```

```python
recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
recommendation_df
```

```python
#user-based recommendation
movies_to_be_recommended = recommendation_df[recommendation_df["weighted_rating"] > 3.5].sort_values("weighted_rating", ascending=False)
```

```python
#Make an item-based suggestion based on the name of the movie that the user has watched with the highest score.
movies_to_be_recommended.merge(movie[["movieId", "title"]])["title"].head()
```

```python
movie_id = rating[(rating["userId"] == random_user) & (rating["rating"] == 5.0)].sort_values(by="timestamp", ascending = False)["movieId"][0:1].values[0]
```

```python
#item-based recommendation
movie_name = movie[movie["movieId"] == movie_id]["title"].values[0]
```

```python
movie_name = user_movie_df[movie_name]
```

```python
movies_from_item_based = user_movie_df.corrwith(movie_name).sort_values(ascending=False)
```
